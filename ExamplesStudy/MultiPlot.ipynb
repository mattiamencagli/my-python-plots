```python
import numpy as np
import matplotlib.pyplot as plt

#F = np.random.normal(loc=1000, scale=50, size=(256, 256))

dim = 256
x = np.arange(0,dim)
y = np.arange(0,dim)
X, Y = np.meshgrid(x,y)
a = 1; b = 1;
F = a*((X-dim*0.5)/b)**2 + a*((Y-dim*0.5)/b)**2

# easy way
# fig0, axs0 = plt.subplots(1, 3, figsize=(15, 5))

# more complex, more control
fig = plt.figure(0, figsize=[12, 5])
gs  = fig.add_gridspec(1,2) #, hspace=0)
axs = gs.subplots()

im1 = axs[0].imshow(F, cmap='viridis', aspect='auto')
axs[0].set_title('F')
fig.colorbar(im1, ax=axs[0])

# +1 to avoid the log to reach -inf
im2 = axs[1].imshow(np.log10(F+1), cmap='viridis', aspect='auto')
axs[1].set_title('Log10 F')
fig.colorbar(im2, ax=axs[1])

plt.tight_layout()
plt.show()
```
