```python
# Lambda examples
```

```python
square = lambda x: x**2
result = square(5)
print(result)
```

```python
# With map and filter
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, numbers))
evens   = list(filter(lambda x: not(x%2), numbers))
odds    = list(filter(lambda x: (x%2), numbers))
print(squared)
print(evens)
print(odds)
```

```python
# Example with dictionaries

data = [{'name': 'Alice', 'age': 28}, {'name': 'Bob', 'age': 21}, {'name': 'Charlie', 'age': 35}]
sorted_data = sorted(data, key=lambda x: x['age'])
print(sorted_data)
```
