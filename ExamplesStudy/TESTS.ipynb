```python
def findMaximumGreatness(arr):
    arr_s = sorted(arr)
    print(arr_s)
    
    # Initialize pointer for the smaller elements to be check
    left_pointer = 0
    
    # Iterate through each element as a potential larger element wrt the left_pointer
    for v in arr_s:
        # If current element is greater than element at left_pointer move the left_pointer forward (counting one success)
        if (v>arr_s[left_pointer]):
            left_pointer += 1
        # print(f"{left_pointer} += {v} > {arr_s[left_pointer]}")
    
    # The final position of left_pointer equals the number of success (new_arr[i]>old_arr[i])
    return left_pointer


array = [4,6,4,2] 
array = [1,2,6,3,2,1] # 4

print(findMaximumGreatness(array))
```

```python
def getMaxSumOfArray(arr1, arr2):
    arr1_is  = sorted(arr1, reverse=True)
    arr2_s = sorted(arr2, reverse=False)
    l = len(arr1)

    summ = 0 
    for i in range(l):
        summ += (i+1)*(arr2_s[i]-arr1_is[i])

    return summ % (7+10**9)

a1 = [1,2,3]
a2 = [10,10,10]
print(getMaxSumOfArray(a1, a2)) # 50
```

```python
def findMinWeight(weights, d):
    # consumo sempre il pi√π pesante che rimane
    for i in range(d):
        weights = sorted(weights, reverse=True)
        weights[0] -= int(weights[0]*0.5)
    return weights

w = [30,20,25]
d = 4
print(findMinWeight(w, d)) # 8,13,10
```
