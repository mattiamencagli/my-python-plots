```python
def findMaximumGreatness(arr):
    arr_s = sorted(arr)
    print(arr_s)
    
    # Initialize pointer for the smaller elements to be check
    left_pointer = 0
    
    # Iterate through each element as a potential larger element wrt the left_pointer
    for v in arr_s:
        # If current element is greater than element at left_pointer move the left_pointer forward (counting one success)
        if (v>arr_s[left_pointer]):
            left_pointer += 1
        # print(f"{left_pointer} += {v} > {arr_s[left_pointer]}")
    
    # The final position of left_pointer equals the number of success (new_arr[i]>old_arr[i])
    return left_pointer


array = [4,6,4,2] 
array = [1,2,6,3,2,1] # 4

print(findMaximumGreatness(array))
```
