```python
import numpy as np
import matplotlib as mplt
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

#F = np.random.normal(loc=1000, scale=50, size=(256, 256))

dim = 256
x = np.arange(0,dim)
y = np.arange(0,dim)
X, Y = np.meshgrid(x,y)
a = 1; b = 1;
F = a*((X-dim*0.5)/b)**2 - a*((Y-dim*0.5)/b)**2

#custom colorbar
vmax=np.amax(F)
vmin=np.amin(F)
norm = mplt.colors.Normalize(vmin,vmax)
colors= [[norm(vmin), "darkblue"],
         [norm(-1000), "lightblue"],
         [norm(-1000), "yellow"],
         [norm(+1000), "yellow"],
         [norm(+1000), "lightgreen"],
         [norm(vmax), "green"]]
cmap = mplt.colors.LinearSegmentedColormap.from_list("", colors)


#custom format on colorbar with a lot of options
def fmt(x, pos):
    a, b = '{:.2e}'.format(x).split('e')
    return r'${} \times 10^{{{}}}$'.format(a, int(b))

plt.figure(0)
plt.contourf(F,255,cmap=cmap)
plt.title('example title')
flag = False
if flag:
    plt.colorbar(format=ticker.FuncFormatter(fmt))
else:
    #force scientific notation
    cbar = plt.colorbar()
    cbar.formatter.set_powerlimits((0, 0))
    cbar.formatter.set_useMathText(True)

save = False
if (save==True): plt.savefig('example.png', format='png',dpi=1000,bbox_inches='tight')
```
