```python
# Examples of Match Case

# simple
def http_status(status):
    match status:
        case 200:
            return "OK"
        case 404:
            return "Not Found"
        case 500:
            return "Server Error"
        case _:
            return "Unknown status"
print(http_status(200))  
print(http_status(404))  
print(http_status(999))  
print()

# with tuples (valid also for dict or lists or whatever)
def describe_point(point):
    match point:
        case (0, 0):
            print("Origin")
        case (0, y):
            print(f"Y axis at {y}")
        case (x, 0):
            print(f"X axis at {x}")
        case (x, y):
            print(f"Point at ({x}, {y})")
        case _:
            print("Not a point")
describe_point((0, 0))  
describe_point((0, 5))  
describe_point((3, 0))  
describe_point((3, 4))  
describe_point('banana')
print()

# with custom objects and if conditions
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
def handle_point(point):
    match point:
        case Point(x=0, y=0):
            print("Origin")
        case Point(x=0, y=y):
            print(f"Y axis at {y}")
        case Point(x=x, y=0):
            print(f"X axis at {x}")
        case Point(x=x, y=y) if x==y: # PAY ATTENTION TO THE ORDER! this must come before the next one, otherwise it will never trigger
            print(f"Point on the bisector at ({x}, {y})")
        case Point(x=x, y=y):
            print(f"Point at ({x}, {y})")
        case _:
            print("Not a point")
handle_point(Point(0, 0))
handle_point(Point(0, 5))
handle_point(Point(3, 0))
handle_point(Point(3, 4))
handle_point(Point(3, 3))
describe_point('banana') 
print()

# coupled with if conditions
def check_num(value):
    match value:
        case x if x < 0:
            print("Negative")
        case x if x == 0:
            print("Zero")
        case x if x > 0:
            print("Positive")
print(check_num(0))  
print(check_num(404))  
print(check_num(-999))  
print()
```
