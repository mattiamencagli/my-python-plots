```python
def primes(limit):
    i = 2
    while i < limit:
        for j in range(2, i):
            if i % j == 0:
                break
        else: # this run if the break do not trigger
            yield i
        i += 1

print(list(primes(20)))
```

```python
# version which makes sense

def primes(limit):
    i = 2
    while i < limit:
        prime_found = True;
        for j in range(2, i):
            if i % j == 0:
                prime_found = False;
                break
        if(prime_found):
            yield i
        i += 1

print(list(primes(20)))
```
