```python
#Iteratable class
class Fibonacci:
    
    def __init__(self, n):
        self.a = 0
        self.b = 1
        self.n = n
        self.count = 0
        
    def __iter__(self):
        return self
        
    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        prt = self.a
        self.count += 1
        self.a, self.b = self.b, self.a + self.b
        return prt

for i, f in enumerate(Fibonacci(15)):
    print(f"{i}\t: {f}")
```
