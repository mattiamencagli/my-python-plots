```python
# example of "unpack operator" (*) for slicing lists
start, *middle, end_1, end = [1, 2, 3, 4, 5]
print(start)
print(middle)
print(end_1)
print(end)

# example of "unpack operator" (*) for tuples
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")
(green, yellow, *red) = fruits
print(green)
print(yellow)
print(red)
```

```python
# * for unpacking arguments
def sum_all(*args):
    return sum(args)

print(sum_all(1, 2, 3))  
```

```python
# ** for packing a dictionary
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name='Alice', age=30)

# ** for passing any arguments in a function
def greet(**kwargs):
    if 'name' in kwargs:
        print(f"Hello, {kwargs['name']}!")
    else:
        print("Hello, stranger!")
greet(name='Alice')
greet(age=30)
```
