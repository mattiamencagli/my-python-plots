```python
def xorhex(hex1, hex2):
    # Conversione da esadecimale a decimale
    dec1 = int(hex1, 16)
    dec2 = int(hex2, 16)
    xor_result = dec1 ^ dec2
    print(f"Primo valore:  esadecimale={hex1}, decimale={dec1}, binario={bin(dec1)[2:].zfill(8)}")
    print(f"Secondo valore: esadecimale={hex2}, decimale={dec2}, binario={bin(dec2)[2:].zfill(8)}")
    print(f"Risultato XOR: esadecimale={hex(xor_result)[2:]}, decimale={xor_result}, binario={bin(xor_result)[2:].zfill(8)}")
    print()


xorhex('7d','20')
xorhex('7e','20')

```
