```python
def estrai_colonne(input_file, output_file, colonne_da_tenere):
    with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
        for line in infile:
            colonne = line.strip().split()  # separo le colonne
            colonne_selezionate = [colonne[i] for i in colonne_da_tenere if i < len(colonne)]
            outfile.write(' '.join(colonne_selezionate) + '\n')

# Esempio di utilizzo
path = 'C:/Users/Mattia Mencagli/Documents/kcf/TRACKER/results/Trasposizione/fromClau/Type2_256/'
input_file = path+'legacyAvtLog.txt'
output_file = path+'ClauStyle.txt'
colonne_da_tenere = [0, 8, 9]

estrai_colonne(input_file, output_file, colonne_da_tenere)
```
