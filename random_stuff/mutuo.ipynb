```python
import matplotlib.pyplot as plt

def calc_mutuo(C, i_annuo, n_anni):
    # Calcolo del tasso mensile e del numero totale di rate
    i_mensile = i_annuo / 12
    N = n_anni * 12
    
    # Calcolo della rata mensile
    R = C * (i_mensile * (1 + i_mensile)**N) / ((1 + i_mensile)**N - 1)
    I = (R*N-C)
    
    # Inizializzazione delle liste per memorizzare le quote di capitale e interessi
    quote_capitale = []
    quote_interessi = []
    
    # Calcolo delle quote di capitale e interessi per ogni rata
    debito_residuo = C
    for mese in range(1, N + 1):
        interessi = debito_residuo * i_mensile
        capitale = R - interessi
        quote_capitale.append(capitale)
        quote_interessi.append(interessi)
        debito_residuo -= capitale

    print("Capitale = %d"%C)
    print("interesse annuo = %1.1f%%"%(i_annuo*100))
    print("n_anni = %d"%n_anni)
    
    # Visualizzazione dei risultati con un plot
    plt.figure(figsize=(6, 4))
    plt.plot(range(1, N + 1), quote_capitale, label='Quota Capitale', color='blue')
    plt.plot(range(1, N + 1), quote_interessi, label='Quota Interessi', color='orange')
    plt.xlabel('Numero Rata')
    plt.ylabel('Importo (â‚¬)')
    plt.title('Scomposizione della Rata: Quota Capitale e Quota Interessi')
    plt.legend()
    plt.grid(True)
    plt.show()

    print("rata = %1.2f"%R)
    print("interessi totali = %1.2f; [Interessi veri totali: %f%%]"%(I,I/C*100))

# Dati del mutuo
Capitale = 200000
interesse_annuo = 0.03
N_anni = 30

calc_mutuo(Capitale, interesse_annuo, N_anni)
```
