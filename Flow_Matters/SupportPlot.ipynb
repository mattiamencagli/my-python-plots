```python
import numpy as np
import matplotlib.pylab as plt
import matplotlib
import h5py as h5
import time
import os

SAVE=True
save_dir = '/home/matti/Pictures/images_sst/'
N = 1000
_ = plt.figure(2)
_ = plt.hlines(0.0004, 0, N, color='blue', label=r"$P_{xy}$ (Gap=256)")
_ = plt.hlines(0.0002, 0, N, color='green', label=r"$P_{xy}$ (Gap=512)")
_ = plt.hlines(0.000133, 0, N, color='red', label=r"$P_{xy}$ (Gap=768)")
_ = plt.title("Shear stress along the gap. Expected scaling.")
# _ = plt.ylim([-0.0001,0.00055])
_ = plt.ylabel('Shear stress')
_ = plt.xlabel('Gap [Z axis]')
_ = plt.grid(alpha=0.2)
_ = plt.legend()
if SAVE:
    _ = plt.savefig(save_dir+'esempio_scaling'+'.png', bbox_inches='tight', dpi=300, format='png')

fn=0.000415
expected_scal = np.array([fn, fn/2, fn/3])
_ = plt.figure(3)
_ = plt.plot(np.array([256,512,768]), expected_scal,  color='black', linewidth=2, label="Expected scaling (linear)")
_ = plt.plot(np.array([256,512,768]),np.array([fn,0.000438,0.000375]),  color='red', linewidth=2, label="Actual scaling")
_ = plt.title("Shear stress scaling comparison")
# _ = plt.ylim([-0.0001,0.00055])
_ = plt.ylabel('Shear stress')
_ = plt.xlabel('Gap size')
_ = plt.xticks([256,512,768])
_ = plt.grid(alpha=0.2)
_ = plt.legend()
if SAVE:
    _ = plt.savefig(save_dir+'esempio_scaling_meglio'+'.png', bbox_inches='tight', dpi=300, format='png')
```
