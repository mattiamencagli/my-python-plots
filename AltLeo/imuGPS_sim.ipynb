```python
import numpy as np

path = 'C:/Users/Mattia Mencagli/Documents/DATA/imuGPS_csv/'

rate = 20; # Hz
Rt = 6357000 # meters
rad_to_deg = 180/np.pi
deg_to_rad = 1/rad_to_deg

Dt = 1/rate # sec
T = 30; # sec
steps = rate * T;

starting_lat = 43.77925 # deg
starting_lon = 11.24626 # deg
Dlat_m       = 1000 # meters
Dlon_m       = 1000 # meters
starting_alt = 500 # meters
ending_alt   = 2000 # meters

starting_roll  = 0 # deg
ending_roll    = 10 # deg
starting_pitch = -5 # deg
ending_pitch   = 5 # deg
starting_yaw   = 10 # deg
ending_yaw     = 20 # deg

ending_lat = starting_lat + (Dlat_m/Rt)*rad_to_deg
ending_lon = starting_lon + (Dlon_m/Rt)*rad_to_deg
lat = np.linspace(starting_lat,ending_lat,steps) # degree
lon = np.linspace(starting_lon,ending_lon,steps) # degree
alt = np.linspace(starting_alt,ending_alt,steps) # meters

roll  = np.linspace(starting_roll ,ending_roll, steps) # meters
pitch = np.linspace(starting_pitch,ending_pitch,steps) # degree
yaw   = np.linspace(starting_yaw,  ending_yaw,  steps) # degree

Nvel_comp = Rt*deg_to_rad*(ending_lat-starting_lat)/T # m/s
Evel_comp = Rt*deg_to_rad*(starting_lon-starting_lon)/T # m/s
Uvel_comp = Rt*deg_to_rad*(ending_alt-starting_alt)/T # m/s
Nvel = np.zeros((steps)) + Nvel_comp # m/s
Evel = np.zeros((steps)) + Evel_comp # m/s
Uvel = np.zeros((steps)) + Uvel_comp # m/s

# da calcolare per benino
yawT = np.zeros((steps)) + 1 # m/s

for step in range(steps):
    with open(path+'immugps_sim.csv', "a") as f:
        f.write(f"{lat[step]};{lon[step]};{alt[step]};{roll[step]};{pitch[step]};{yaw[step]};{Nvel[step]};{Evel[step]};{Uvel[step]};{yawT[step]}\n")
```
