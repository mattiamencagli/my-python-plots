```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Delta = 20
dLineX = 0.1

azimut = np.pi * 5 / 6
dGround = 70


N = int(2*Delta/dLineX)

XNord = np.zeros((N))
YEast = np.zeros((N))
XNord[0] = dGround*np.cos(azimut) - Delta*np.cos(azimut) - Delta*np.sin(azimut)
YEast[0] = dGround*np.sin(azimut) - Delta*np.sin(azimut) + Delta*np.cos(azimut)

direz = 1;
for i in range(1,N):
    XNord[i] = XNord[i-1] + direz*dLineX*np.sin(azimut)
    YEast[i] = YEast[i-1] - direz*dLineX*np.cos(azimut)
```

```python
plt.figure(0,figsize=[5,5])
plt.plot(XNord[0:int(N/2)],YEast[0:int(N/2)])
plt.plot(XNord[int(N/2):-1],YEast[int(N/2):-1])
plt.plot([0,dGround*np.cos(azimut)],[0,dGround*np.sin(azimut)],color='green',linestyle='--', linewidth=0.5)
plt.hlines(0,-200,200,color='red',linestyle='--', linewidth=0.5)
plt.vlines(0,-200,200,color='red',linestyle='--', linewidth=0.5)
plt.xlim([-100,100])
plt.ylim([-100,100])
plt.grid(alpha=0.3)
```
