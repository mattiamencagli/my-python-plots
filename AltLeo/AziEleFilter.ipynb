```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm as sp
%config InlineBackend.figure_format = 'svg'

path = 'C:/Users/Mattia Mencagli/Documents/DATA/AziEleFilter/'
name = 'azi_ele_filterON_old'
name = 'azi_ele_filterON_new'
name = 'azi_ele_filterON_solved'

data = np.genfromtxt(path+name+'.csv', delimiter=' ')
# ae_rel_FNO = data[:,0:2] # inutile
ae_abs_FNO = data[:,2:4]
ae_abs_FSI = data[:,4:6]
#ae_rel_FNO.shape
```

```python
i = 0; f = len(data[:,0])
def X(i,f): return np.linspace(i,f,f-i)

# i = 10; f = 45
# i = 70; f = 120
# i = 1200; f = 1400
_ = plt.figure(0)
_ = plt.plot(X(i,f), ae_abs_FNO[i:f,0], label = 'FILTER OFF')
_ = plt.plot(X(i,f), ae_abs_FSI[i:f,0], label = 'FILTER ON ')
_ = plt.hlines([-np.pi,np.pi], i, f, color='black', linestyle=':', linewidth=1)
# _ = plt.plot(ae_abs_FSI[i:f,0] - ae_abs_FNO[i:f,0], label = 'FILTER ON - FILTER OFF')
_ = plt.legend()
_ = plt.title('Azimut')
_ = plt.xlabel('Steps')
_ = plt.ylabel('Radians')
_ = plt.savefig(path+'azimut.svg', bbox_inches='tight', format='svg')

# i = 350; f = 400
_ = plt.figure(1)
_ = plt.plot(X(i,f), ae_abs_FNO[i:f,1], label = 'FILTER OFF')
_ = plt.plot(X(i,f), ae_abs_FSI[i:f,1], label = 'FILTER ON ')
# _ = plt.plot(ae_abs_FSI[i:f,1] - ae_abs_FNO[i:f,1], label = 'FILTER ON - FILTER OFF')
_ = plt.legend()
_ = plt.title('Elevation')
_ = plt.xlabel('Steps')
_ = plt.ylabel('Radians')
```
