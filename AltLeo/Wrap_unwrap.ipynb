```python
import numpy as np

def rewrap_easy(a):
    ret = a
    while( ret < -2*np.pi ):
        ret += 2*np.pi
    while( ret > +2*np.pi ):
        ret -= 2*np.pi
    return ret

def rewrap_furbo(a):
    return a - int(a/(2*np.pi))*2*np.pi

# angles = -2.1 * np.pi
angles = np.linspace(-7*np.pi, 9*np.pi, 20)
for angle in angles:
    wrap_angle = rewrap_easy(angle)
    print("easy  : {} PI".format(wrap_angle/np.pi))
    wrap_angle = rewrap_furbo(angle)
    print("furbo  : {} PI".format(wrap_angle/np.pi))
    print()
```
