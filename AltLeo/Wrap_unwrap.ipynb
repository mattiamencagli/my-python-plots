```python
import numpy as np

def rewrap_easy(a):
    ret = a
    while( ret < -2*np.pi ):
        ret += 2*np.pi
    while( ret > +2*np.pi ):
        ret -= 2*np.pi
    return ret

def rewrap_furbo(a):
    return a - int(a/(2*np.pi))*2*np.pi

# angles = -2.1 * np.pi
angles = np.linspace(-30*np.pi, 30*np.pi, 10000)
for angle in angles:
    wrap_angle_easy = rewrap_easy(angle)
    wrap_angle_furbo = rewrap_furbo(angle)
    if (np.abs(wrap_angle_furbo-wrap_angle_easy)>1e-9):
        print("ERROR : angle = {} PI".format(angle/np.pi))
        print("easy  : {} PI".format(wrap_angle_easy/np.pi))
        print("furbo  : {} PI".format(wrap_angle_furbo/np.pi))
        print()
        
```
