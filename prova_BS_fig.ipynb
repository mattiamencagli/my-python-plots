```python
import numpy as np
import matplotlib.pylab as plt
```

```python
direc = '/home/matti/programming/prova_BS_ARchain/cmake-build-debug/'
name = 'orbite_AR_TZU_precise.dat'
# name = 'orbite_AR_TZU.dat'
# name1 = 'orbite_AR_TZU_BELLINE_07.dat'
f = open(direc + name)
file=f.readlines()
f.close()
N=int(file[0].rstrip())
file.pop(0)
o = np.loadtxt(file)
# x, y, z, vx, vy, vz, ax, ay, az, err, time
print(N)
print(o.shape)
```

```python
n=2
fs = 'xx-large'
lab = ['b1', 'b2', 'b3', 'b4', 'b5', 'b6', 'b7', 'b8', 'b9', 'b10']
col = ['blue', 'red', 'green','purple', 'orange']
%matplotlib inline
#%matplotlib widget
```

```python
# f = 48845 #trovato a mano con tanto ammmore
f = int(1091874 * 0.25) #trovato a mano con tanto ammmore
div = 1000;
#_ = plt.figure(0, figsize=[n*6.4,n*4.8])
_ = plt.figure(0, figsize=[n*4.8*1.6,n*4.8])
for i in range(N):
    _ = plt.scatter(o[i:f:N*div,0],o[i:f:N*div,1], label=lab[i], lw=1, color=col[i])
_ = plt.scatter(o[0:N,0],o[0:N,1], color='black', marker='s',lw=3)
_ = plt.xlim(-1.8,1.8)
_ = plt.ylim(-0.3,0.3)
#_ = plt.legend(fontsize=fs)
_ = plt.xticks(fontsize=fs)
_ = plt.yticks(fontsize=fs)
_ = plt.xlabel('X', fontsize=fs)
_ = plt.ylabel('Y',fontsize=fs)
_ = plt.savefig('./3body_8.png', bbox_inches='tight', format='png')
#_ = plt.savefig('./pitagorean_orb.png', bbox_inches='tight', format='png')
#_ = plt.savefig('./pitagorean_orb_belline.png', bbox_inches='tight', format='png')
```
